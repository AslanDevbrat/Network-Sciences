```python
import numpy as np
import matplotlib.pyplot as plt
from math import e

S = np.arange(0,1, 0.0001) # Creating evenly spaced array from 0 to 1 with step size 10000

y1 = 1-e**(-0.5*S)
y2 = 1-e**(-1.0*S)
y3 = 1-e**(-1.5*S)
```

```python
plt.rcParams["figure.figsize"] = (8,8)
plt.plot(S,S,label='y=S',ls='--')
plt.plot(S,y1, label = 'c = 0.5')
plt.plot(S,y2,label = 'c = 1.')
plt.plot(S,y3,label = 'c = 1.5')
plt.axvline(x=0.581)
plt.ylabel('y',fontsize='xx-large')
plt.xlabel('S',fontsize='xx-large')
plt.grid()
plt.legend(loc='best',fontsize='xx-large')
plt.show()
```

```python
c = -np.log(1-S)/S

plt.plot(c,S,label='y=S',ls='--') 
plt.xlim([0,4])
plt.ylim([0,1])
plt.xlabel('Average degree c',fontsize='xx-large')
plt.ylabel('Size of giant component S', fontsize='xx-large')
```
