```python
import numpy as np
import matplotlib.pyplot as plt
from math import e

S = np.arange(0,1, 0.0001) # Creating evenly spaced array from 0 to 1 with step size 10000

y1 = 1-e**(-0.5*S)       # Creating an array to hold y values for c = 0.5
y2 = 1-e**(-1.0*S)       # Creating an array to hold y values for c = 1
y3 = 1-e**(-1.5*S)       # Creating an array to hold y values for c = 1.5
```

```python
plt.rcParams["figure.figsize"] = (8,8)        #making the plot bigger  
plt.plot(S,S,label='y=S',ls='--')                # Ploting y=S
plt.plot(S,y1, label = 'c = 0.5')                  # Ploting y for c = 0.5
plt.plot(S,y2,label = 'c = 1.')                     # Ploting y for c = 1.0
plt.plot(S,y3,label = 'c = 1.5')                   # Ploting y for c = 1.5
plt.axvline(x=0.581)
plt.ylabel('y',fontsize='xx-large')            #Increasing the font size of the label verticle axis 
plt.xlabel('S',fontsize='xx-large')           #Increasing the font size of the label Horizontal axis 
plt.grid()                                                   # Showing the grid 
plt.legend(loc='best',fontsize='xx-large')
plt.show()
```

Here I have used its 1st equation which involves Lambert-W-function.


```python
c = np.arange(0,3,0.0001)
from scipy.special import lambertw                # importing Library to use lambert w function
import sys
s=[]                                                                 # array to store mean size of the graph for different c values

"""
    This loop is on c array which contain evenly spaced 10000 number 
    between 0 and 3
    
    In this loop if the value in the c is less than 1 then 
        s = 1/(1-c)
    otherwise
        Here We first have tocalculate S then using this S
        we calculate s using the following formule
        S = 1+ (lambertWFuntion(-c *e^(-c) ))/c
        s = 1/(1-c+c*S)
        
"""

for i,element in enumerate(c):
    if element<1:
        s.append(1/(1-element))                          # s = 1/(1-c)
    else:
        S = 1+lambertw(-element*e**(-element))/element     # First calculating S
        s.append( 1/(1-element+element*S))                           # using S to calculate average size <s>

plt.xlim([0,3])                                                                      # limiting x axis from 0 to 3
plt.ylim([0,8])                                                                      # limiting y axis from 0 to 8
plt.axvline(x=1,ls='--')                                                         # Creating a verticle line at x=1 
plt.plot(c,s)                                                                          # ploting mean degree vs mean size
plt.grid()
plt.xlabel('Mean degree c',fontsize='xx-large')
plt.ylabel('Size ', fontsize='xx-large')
```
