```python
import numpy as np
"""
    This function check weather a graph is directed or Undirected.
    To check this we need to only check if the graph is symmetric or
    not. TO check this I have just calculated matrix's transpose and 
    subtracted it with the original matrix. if all the element will become 
    0 the it is symmetric otherwise not.
"""

def check_symmetric(a, tol=0):
    return np.all(np.abs(a-a.T) == tol)
#check_symmetric(np.array(adj4))
```

```python
def test_directed(adj_matrix_list):
    for i,matrix in enumerate(adj_matrix_list):
        result = check_symmetric(np.array(matrix))
        if result:
            print(f"Graph {i+1} is Undirected")
        else:
            print(f"Graph {i+1} is Directed")
```

## Pseudocode for checking cycles in the Directed/Undirected Graph 

    L  Empty list that will contain the sorted elements
    S  Set of all nodes with no incoming edge

    while S is not empty do
        remove a node n from S
        add n to L
        for each node m with an edge e from n to m do
            remove edge e from the graph
                if m has no other incoming edges then
                    insert m into S

    if graph has edges then
        print Cyclic Graph
    else 
        print Acyclic
